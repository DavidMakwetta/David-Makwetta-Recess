### Using pd.read_csv() to load the file

### Plotting A Scatter graph 

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
scatter = plt.scatter(df['Maxpulse'], df['Calories'], c=df['Calories'], cmap='plasma', edgecolor='black')

plt.title('Maxpulse vs Calories Burned')
plt.xlabel('Maxpulse')
plt.ylabel('Calories')
plt.colorbar(scatter, label='Calories')  # Show color scale
plt.grid(True)
plt.show()
```

### Plotting A line Graph 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(df.index, df['Maxpulse'], label='Maxpulse', color='blue', marker='o')
plt.plot(df.index, df['Calories'], label='Calories', color='red', marker='x')

plt.title('Maxpulse and Calories Over Data Points')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
